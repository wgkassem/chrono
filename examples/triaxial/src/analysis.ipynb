```python
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
output_notebook()

df = pd.read_csv("~/build_chrono/examples/triaxial/mixer_output/fticks.csv", header=0)

df = df.rename(columns=lambda x: x.strip())
print(df.head())
print("loaded")
#iikjhbkjb
p = figure()
p.xaxis.axis_label = "Time (s)"
p.yaxis.axis_label = "Pressure (kPa)"
p.line(df['curr_time'], df['Prr']*0.0001, legend_label="Radial Pressure", line_width=2, color='red')
p.line(df['curr_time'], df['Pzz']*0.0001, legend_label="Axial Pressure", line_width=2, color='blue')
p.legend.location = "bottom_right"
show(p)
```
